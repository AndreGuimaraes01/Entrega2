## **2. BFS, Breadth-First Search(2,0)**

```python
# A busca em largura explora o grafo nível por nível. É um algoritmo de busca cega, pois não utiliza informações sobre a distância até o objetivo, apenas a estrutura do grafo.

from collections import deque

def bfs(grafo, inicio, destino):
    fila = deque([(inicio, [inicio])])
    visitados = set()

    while fila:
        (no_atual, caminho) = fila.popleft()
        if no_atual == destino:
            return caminho

        if no_atual not in visitados:
            visitados.add(no_atual)
            for vizinho in grafo[no_atual]:
                if vizinho not in visitados:
                    fila.append((vizinho, caminho + [vizinho]))
    return None
```

```python
# Seus Códigos aqui intercalados de células de texto com comentários
#Nesta seção, realizaremos testes utilizando o algoritmo de Busca em Largura (BFS).
#O objetivo é encontrar um caminho entre cidades das regiões Norte e Nordeste. Como o BFS é um algoritmo de busca cega, ele encontrará o caminho com o menor número de conexões (nós) entre a origem e o destino, independentemente da distância em quilômetros.

# Teste do algoritmo BFS
origem = 'Manaus'
destino = 'Recife'

caminho_bfs = bfs(grafo_nne, origem, destino)

print(f"Resultado da busca BFS de {origem} para {destino}:")
if caminho_bfs:
    print(f"Caminho encontrado: {' -> '.join(caminho_bfs)}")
    print(f"Número de cidades no caminho: {len(caminho_bfs)}")
else:
    print("Nenhum caminho encontrado.")
```
